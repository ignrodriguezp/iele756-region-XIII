# Part 1: Census 2024

En primer lugar, se realiza la carga inicial de los datos del Censo 2024 usando la librería pandas.

Los datos fueron utilizados en formato parquet. Luego, se seleccionan solo las columnas específicas que usaremos.

```python
import pandas as pd

persona = pd.read_parquet(
"personas_censo2024.parquet",
columns=["region", "comuna", "sexo", "edad",
"p27_nacionalidad", "p27_nacionalidad_rec",
"escolaridad", "sit_fuerza_trabajo"],
)
```

El siguiente fragmento de código realiza una inspección básica de salud de los datos para asegurar que se cargaron correctamente.


```python
#1.2
#Inspeccionamos parámetros y los mostramos en pantalla
print("Filas; Columnas")
print(persona.shape)
print("Tipo de datos:")
print(persona.dtypes)
print("Primeras 10 filas:")
print(persona.head(10))
print("Valores nulos:")
print(persona.info())
```

De la inspección anterior se puede concluir que tabla en total tiene 15.360.291 filas y 8 columnas. En la clasificación de tipo de datos las columnas region, comuna, sexo y edad son enteros, mientras que p27_nacionalidad, p27_nacionalidad_rec, escolaridad y sit_fuerza_trabajo son números reales.

A continuación filtramos la región seleccionada, en este caso la Región Metropolitana (13) y se imprimen la cantidad de filas en esta región (740.899).

```python
#1.3 Filtrar región
my_region = persona[persona["region"] == 10]
print(f"Rows in my region: {len(my_region):,}")
```

Luego, se cuenta cuántas personas hay en cada categoría de nacionalidad dentro de la región filtrada.

```python
foreign = my_region["p27_nacionalidad_rec"].value_counts()
print(foreign)
```

Es posible concluir que en la Región Metropolitana 712.338 personas representan a los nacidos en Chile, 25.061 personas representan a los nacidos en el extranjero y 3.499 personas son los datos faltantes.

```python
#1.4 Nacionalidad
foreign = my_region["p27_nacionalidad_rec"] \
.value_counts(normalize=True)
print(f"% foreign-born: {foreign.get(2, 0):.1%}")
```

Finalmente, según los datos cargados, el 3.4% de las personas de la tabla filtrada de la Región Metropolitana son nacidas en el extranjero.

# Part 2: ENO — Notifiable Diseases
### 2.1 Carga del dataset ENO
En primer lugar, cargamos el dataset ENO e imprimimos el número total de filas y los nombres de las columnas


```python
import pandas as pd

eno = pd.read_csv(
    "20241218_base_eno_final.csv",
    sep=";", encoding="utf-8-sig",
    low_memory=False)
print(f"Total rows: {len(eno)}")
print(eno.columns.tolist())
```

### 2.2 Filtrado por región

El dataset ENO contiene una columna llamada region, en donde se filtraron los datos para considerar únicamente los registros correspondientes a la Región Metropolitana de Santiago.  
Luego se imprime el número de observaciones presentes en esta región.

```python
eno_region = eno[eno["region"] == "Región Metropolitana de Santiago"]

print(f"Rows in my region: {len(eno_region)}")
```

### 2.3 Número de notificaciones por año

Aora, se calcula el número de notificaciones de enfermedades por año , luego, los resultados se presentan mediante un gráfico de barras para visualizar la evolución temporal de los casos.

```python
eno_region["anho_notificacion"].value_counts().sort_index()
```

```python
notificaciones_por_anio = eno_region["anho_notificacion"].value_counts().sort_index()

import matplotlib.pyplot as plt
notificaciones_por_anio.plot(kind="bar", figsize=(10,5))

plt.title("Notificaciones ENO por año - Región Metropolitana")
plt.xlabel("Año")
plt.ylabel("Número de notificaciones")
plt.show()
```

### 2.4 Enfermedades más notificadas

Se identifican las cinco enfermedades más frecuentemente notificadas de la Region Metropolitana, los resultados se visualizan mediante un gráfico de barras horizontal a modo de comparar fácilmente la frecuencia de cada enfermedad.

```python
eno_region["ENO"].value_counts().head(5)
```

```python
top5_enfermedades = eno_region["ENO"].value_counts().head(5)
top5_enfermedades.plot(kind="barh", figsize=(8,5))

plt.title("Top 5 enfermedades notificadas - Región Metropolitana")
plt.xlabel("Número de casos")
plt.ylabel("Enfermedad")
plt.show()
```

### 2.5 Distribución de nacionalidad

Finalmente, se busca comparar la distribucion de nacionalidad de las personas asociadas a los registros de enfermedades, para esto se cuentan las ocurrencias de cada categoría, luego, a modo de simplificar la comparación, estos datos se grafican en un grafico de barras.

```python
nacionalidad_dist = eno_region["nacionalidad"].value_counts()

print(nacionalidad_dist)
```

```python
nacionalidad_dist.head(10).plot(kind="bar", figsize=(10,5))

plt.title("Distribución de nacionalidad - Región Metropolitana")
plt.xlabel("Nacionalidad")
plt.ylabel("Número de casos")
plt.xticks(rotation=45)

plt.show()
```

#Parte 3: GRD — Hospital Discharges
### 3.1 Carga del dataset GRD

En esta sección se carga el dataset GRD correspondiente al año 2024, el cual contiene información sobre egresos hospitalarios.  
Debido al gran tamaño del archivo, se seleccionan únicamente las columnas necesarias para el análisis mediante el parámetro usecols.  
Luego, se imprime el número total de registros cargados.

```python
import pandas as pd

cols = [ "COMUNA", "NACIONALIDAD", "SEXO", "DIAGNOSTICO1",
    "FECHA_INGRESO", "FECHAALTA",
    "IR_29301_SEVERIDAD", "IR_29301_COD_GRD"]

grd = pd.read_csv(
    "GRD_PUBLICO_2024.txt",
    sep="|",
    usecols=cols,
    low_memory=False,
    encoding="latin-1")
print(f"Total discharges: {len(grd):,}")
```

### 3.2 Filtrado por región mediante comunas

El dataset GRD no incluye directamente una columna de región, por lo que se utiliza la variable COMUNA para identificar los registros correspondientes a la región seleccionada, en este caso, Región Metropolitana.
Se define una lista de comunas pertenecientes a esta región y se filtran los datos.
Luego se imprime el número de egresos hospitalarios correspondientes a la Región Metropolitana.

```python
my_comunas = [
"ALHUE","BUIN","CALERA DE TANGO","CERRILLOS","CERRO NAVIA","COLINA",
"CONCHALI","CURACAVI","EL BOSQUE","EL MONTE","ESTACION CENTRAL",
"HUECHURABA","INDEPENDENCIA","ISLA DE MAIPO","LA CISTERNA",
"LA FLORIDA","LA GRANJA","LA PINTANA","LA REINA","LAMPA",
"LAS CONDES","LO BARNECHEA","LO ESPEJO","LO PRADO","MACUL",
"MAIPU","MARIA PINTO","MELIPILLA","ÑUÑOA","PADRE HURTADO",
"PAINE","PEDRO AGUIRRE CERDA","PEÑAFLOR","PEÑALOLEN","PIRQUE",
"PROVIDENCIA","PUDAHUEL","PUENTE ALTO","QUILICURA","QUINTA NORMAL",
"RECOLETA","RENCA","SAN BERNARDO","SAN JOAQUIN","SAN JOSE DE MAIPO",
"SAN MIGUEL","SAN PEDRO","SAN RAMON","SANTIAGO","TALAGANTE",
"TILTIL","VITACURA"
]

grd_region = grd[grd["COMUNA"].isin(my_comunas)]

print(len(grd_region))
```

### 3.3 Unión con la tabla de diagnósticos CIE-10

Los diagnósticos en el dataset GRD se encuentran codificados según el estándar CIE-10.  
Para obtener la descripción de cada diagnóstico, se realiza una unión entre el dataset GRD y la tabla CIE-10 utilizando el código de diagnóstico como clave.  
De esta forma se agregan columnas descriptivas que permiten interpretar los diagnósticos registrados.

```python
cie10 = pd.read_excel("CIE-10.xlsx", sheet_name="CIE 10")

grd_region = grd_region.merge(
    cie10[["Código", "Descripción", "Capítulo"]],
    left_on="DIAGNOSTICO1",
    right_on="Código",
    how="left")
```

### 3.4 Diagnósticos más frecuentes

Finalmente, se identifican los cinco diagnósticos hospitalarios más frecuentes de la Region Metropolitana tras el conteo de las ocurrencias de cada descripción de diagnóstico y se seleccionan los cinco más comunes.  
Los resultados se visualizan mediante un gráfico de barras horizontal que facilita la comparación entre diagnósticos.

```python
grd_region["Descripción"].value_counts().head(5)
```

```python
top5_diag = grd_region["Descripción"].value_counts().head(5)

top5_diag.plot(kind="barh", figsize=(8,5))

import matplotlib.pyplot as plt
plt.title("Top 5 diagnósticos hospitalarios - Región Metropolitana")
plt.xlabel("Número de casos")
plt.ylabel("Diagnóstico")

plt.show()
```
