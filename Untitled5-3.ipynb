### A.0 Data Loading & Cleaning 

#### A.0.1 Load the ENO dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display

eno_cols = [
    "ENO",
    "anho_notificacion",
    "region",
    "codigo_comuna_residencia",
    "nacionalidad",
    "sexo",
    "grupo_edad",
    "nombre_instruccion",
    "cie_10_diagnostico",
    "diagnostico",
    "pais_contagio"
]

eno = pd.read_csv(
    "/Users/ignaciorodriguezperez/Desktop/Preparacion_datos/20241218_base_eno_final.csv",
    sep=";",
    encoding="utf-8-sig",
    usecols=eno_cols
)

print("Shape ENO:", eno.shape)
print("\nInfo ENO:")
print(eno.info())
```

#### A.0.2 Filter the comunas

```python
MY_COMUNAS = [13102, 13105, 13302]  # Cerrillos, Estación Central, Lampa

my_comunas_str = [str(c) for c in MY_COMUNAS]

eno_com = eno[eno["codigo_comuna_residencia"].isin(my_comunas_str)].copy()

print(f"\nRows in my comunas: {len(eno_com):,}")

# Filas anonimizadas en TODO el dataset
anon_rows = (eno["codigo_comuna_residencia"] == "*****").sum()
anon_pct = anon_rows / len(eno) * 100

print(f"Rows anonymized in full ENO dataset: {anon_rows:,}")
print(f"Percent anonymized in full ENO dataset: {anon_pct:.2f}%")
```

#### A.0.3 Clean the nationality variable

```python
# Distribución de nacionalidad en las comunas filtradas
nat_dist = eno_com["nacionalidad"].value_counts(dropna=False).reset_index()
nat_dist.columns = ["nacionalidad", "count"]
nat_dist["pct"] = nat_dist["count"] / len(eno_com) * 100

print("Distribution of nacionalidad in filtered ENO data:")
print(nat_dist)

# Cantidad de filas "Desconocido"
n_desconocido = (eno_com["nacionalidad"] == "Desconocido").sum()
pct_desconocido = n_desconocido / len(eno_com) * 100

print(f"\nRows with nacionalidad == 'Desconocido': {n_desconocido:,}")
print(f"Percent of filtered ENO data: {pct_desconocido:.2f}%")

# Base para análisis por nacionalidad: excluir 'Desconocido'
eno_com_nat = eno_com[eno_com["nacionalidad"] != "Desconocido"].copy()

print(f"\nRows kept for nationality-specific analyses: {len(eno_com_nat):,}")
print(f"Rows dropped for nationality-specific analyses: {n_desconocido:,}")
```

```python
# Ordenar de mayor a menor
nat_dist_sorted = nat_dist.sort_values(by="count", ascending=True)

plt.figure(figsize=(8,5))

plt.barh(
    nat_dist_sorted["nacionalidad"],
    nat_dist_sorted["count"]
)

plt.xlabel("Number of notifications")
plt.ylabel("Nationality")
plt.title("Distribution of Nationality in ENO (My Comunas)")

# Agregar etiquetas de porcentaje
for i, (count, pct) in enumerate(zip(nat_dist_sorted["count"], nat_dist_sorted["pct"])):
    plt.text(count, i, f" {pct:.1f}%", va='center')

plt.tight_layout()
plt.show()
```

#### A.0.4 Report the time span

```python
# Asegurar que el año sea numérico
eno_com["anho_notificacion"] = pd.to_numeric(eno_com["anho_notificacion"], errors="coerce")

# Rango de años
year_min = eno_com["anho_notificacion"].min()
year_max = eno_com["anho_notificacion"].max()

print(f"Time span in filtered ENO data: {int(year_min)} - {int(year_max)}")

# Notificaciones por año
eno_yearly = (
    eno_com["anho_notificacion"]
    .value_counts()
    .sort_index()
    .reset_index()
)

eno_yearly.columns = ["year", "notifications"]

print("\nNotifications per year:")
print(eno_yearly)
```

Al cargar el dataset original, este cuenta con **333.300 registros** a nivel nacional. Luego de filtrar por nuestras comunas asignadas (Cerrillos, Estación Central y Lampa), nos quedamos con **2.526 notificaciones**. 

Es importante destacar que en el dataset completo existe un **44.47% de registros anonimizados** (con el valor `*****` en el código de comuna), lo que significa que estamos perdiendo un volumen considerable de datos espaciales. 

Respecto a la nacionalidad en nuestra muestra, un **52.3%** corresponde a chilenos, un **12.1%** a extranjeros y un **35.6%** a nacionalidad "Desconocida". Dado el alto volumen de datos faltantes en esta variable, se decidió excluir los 900 casos "Desconocidos" para los futuros análisis específicos de nacionalidad, conservando una base limpia de **1.626 registros**.

### A.2.2 Disease profiles by nationality

```python
top5_diseases = (
    eno_com["ENO"]
    .value_counts()
    .head(5)
    .index
    .tolist()
)

print("Top 5 diseases:", top5_diseases)

# Filtrar base sin "Desconocido" a esas top 5 enfermedades
eno_top5_nat = eno_com_nat[eno_com_nat["ENO"].isin(top5_diseases)].copy()

# Contar notificaciones por enfermedad y nacionalidad
eno_top5_nat_counts = (
    eno_top5_nat
    .groupby(["ENO", "nacionalidad"])
    .size()
    .reset_index(name="notifications")
)

print("\nCounts by disease and nationality:")
print(eno_top5_nat_counts)

# Pivot para gráfico
eno_top5_nat_pivot = (
    eno_top5_nat_counts
    .pivot(index="ENO", columns="nacionalidad", values="notifications")
    .fillna(0)
)

# Asegurar orden de enfermedades según top 5
eno_top5_nat_pivot = eno_top5_nat_pivot.reindex(top5_diseases)

# Si falta una columna, crearla en 0
for col in ["Chile", "Extranjero"]:
    if col not in eno_top5_nat_pivot.columns:
        eno_top5_nat_pivot[col] = 0

# Calcular shares dentro de cada enfermedad
eno_top5_nat_share = eno_top5_nat_pivot.div(eno_top5_nat_pivot.sum(axis=1), axis=0) * 100

print("\nShare by disease and nationality (%):")
print(eno_top5_nat_share.round(2))
```

```python
x = np.arange(len(eno_top5_nat_share.index))
width = 0.35

plt.figure(figsize=(10, 6))

plt.bar(
    x - width/2,
    eno_top5_nat_share["Chile"],
    width=width,
    label="Chile"
)

plt.bar(
    x + width/2,
    eno_top5_nat_share["Extranjero"],
    width=width,
    label="Extranjero"
)

plt.xticks(x, eno_top5_nat_share.index, rotation=20, ha="right")
plt.ylabel("Share of notifications (%)")
plt.xlabel("Disease")
plt.title("Nationality Share for Top 5 ENO Diseases")
plt.legend()
plt.tight_layout()
plt.show()
```

## A.4 Comuna-Level ENO Summary Table

```python
def build_eno_summary(group):
    codigo = int(group.name)

    eno_total = len(group)
    eno_chilean = (group["nacionalidad"] == "Chile").sum()
    eno_foreign = (group["nacionalidad"] == "Extranjero").sum()
    eno_desconocido = (group["nacionalidad"] == "Desconocido").sum()

    top3 = (
        group["ENO"]
        .value_counts()
        .head(3)
        .index
        .tolist()
    )
    eno_top3_diseases = ", ".join(top3)

    return pd.Series({
        "codigo_comuna": codigo,
        "eno_total": eno_total,
        "eno_chilean": eno_chilean,
        "eno_foreign": eno_foreign,
        "eno_desconocido": eno_desconocido,
        "eno_top3_diseases": eno_top3_diseases
    })

eno_summary = (
    eno_com
    .groupby("codigo_comuna_residencia")
    .apply(build_eno_summary)
    .reset_index(drop=True)
)

# Cargar resumen de Tarea 1
t1 = pd.read_csv("/Users/ignaciorodriguezperez/Desktop/Preparacion_datos/output/tarea1_comuna_summary_final.csv")
t1["comuna"] = t1["comuna"].astype(int)


# Merge con Tarea 1
eno_summary = eno_summary.merge(
    t1[["comuna", "nombre_comuna", "pop_total"]],
    left_on="codigo_comuna",
    right_on="comuna",
    how="left"
)

# Calcular tasa por 10.000 habitantes
eno_summary["eno_rate_per_10k"] = (
    eno_summary["eno_total"] / eno_summary["pop_total"] * 10000
).round(2)

# Seleccion de columnas y orden
eno_summary = eno_summary[
    [
        "codigo_comuna",
        "nombre_comuna",
        "eno_total",
        "eno_chilean",
        "eno_foreign",
        "eno_desconocido",
        "eno_top3_diseases",
        "eno_rate_per_10k"
    ]
]

# Ordenar por comuna
eno_summary = eno_summary.sort_values("codigo_comuna").reset_index(drop=True)


display(
    eno_summary.style
    .format({
        "eno_total": "{:,.0f}",
        "eno_chilean": "{:,.0f}",
        "eno_foreign": "{:,.0f}",
        "eno_desconocido": "{:,.0f}",
        "eno_rate_per_10k": "{:.2f}"
    })
    .hide(axis="index")
)

import os

# Crear carpeta output si no existe
os.makedirs("output", exist_ok=True)

# Guardar CSV
eno_summary.to_csv("output/tarea2_eno_summary.csv", index=False)

print("Archivo guardado en output/tarea2_eno_summary.csv")
```

```python
base = "/Users/ignaciorodriguezperez/Desktop/Preparacion_datos"

grd_cols = [
    "COMUNA",
    "NACIONALIDAD",
    "SEXO",
    "FECHA_NACIMIENTO",
    "FECHA_INGRESO",
    "FECHAALTA",
    "TIPOALTA",
    "DIAGNOSTICO1",
    "DIAGNOSTICO2",
    "IR_29301_SEVERIDAD",
    "IR_29301_MORTALIDAD",
    "IR_29301_COD_GRD",
    "TIPO_INGRESO",
    "ESPECIALIDAD_MEDICA"
]

files = {
    2022: ("GRD_PUBLICO_EXTERNO_2022.txt", "utf-16"),
    2023: ("GRD_PUBLICO_2023.txt", "utf-16"),
    2024: ("GRD_PUBLICO_2024.txt", "latin-1")
} ## Se modificó el encodin de 2022 y 2023 ya que no reconocia latin-1, se descargó en utf-16

frames = []

for year, (file, enc) in files.items():
    path = os.path.join(base, file)

    df = pd.read_csv(
        path,
        sep="|",
        encoding=enc,
        usecols=grd_cols,
        low_memory=False
    )

    df.columns = df.columns.str.strip()
    df["year"] = year
    frames.append(df)

    print(f"Año {year} cargado correctamente con encoding {enc}. Shape: {df.shape}")

grd = pd.concat(frames, ignore_index=True)

print("\nBase GRD cargada correctamente")
print("Shape total:", grd.shape)
print("\nFilas por año:")
print(grd["year"].value_counts().sort_index())
print("\nColumnas:")
print(grd.columns.tolist())
```

### B.0.3 Compute length of stay

```python
# Convertir fechas
grd_com["fecha_ingreso_dt"] = pd.to_datetime(grd_com["FECHA_INGRESO"], errors="coerce")
grd_com["fecha_alta_dt"] = pd.to_datetime(grd_com["FECHAALTA"], errors="coerce")

# Se crea los (length-of-stay)
grd_com["los"] = (grd_com["fecha_alta_dt"] - grd_com["fecha_ingreso_dt"]).dt.days

# Resumen inicial
print("LOS summary before filtering negative values:")
print(grd_com["los"].describe())

print("\nMean LOS:", round(grd_com["los"].mean(), 2))
print("Median LOS:", grd_com["los"].median())
print("Min LOS:", grd_com["los"].min())
print("Max LOS:", grd_com["los"].max())

# Contar filas "los" con negativo
n_negative_los = (grd_com["los"] < 0).sum()

print(f"\nRows with los < 0: {n_negative_los:,}")

# Filtrar errores
grd_com = grd_com[grd_com["los"] >= 0].copy()

print(f"Rows after removing negative LOS: {len(grd_com):,}")

# Resumen final
print("\nLOS summary after filtering negative values:")
print(grd_com["los"].describe())

print("\nMean LOS:", round(grd_com["los"].mean(), 2))
print("Median LOS:", grd_com["los"].median())
print("Min LOS:", grd_com["los"].min())
print("Max LOS:", grd_com["los"].max())
```

Se calculó la estadía hospitalaria como la diferencia entre FECHAALTA y FECHA_INGRESO, medida en días.

Se obtuvieron 33,126 observaciones válidas de los 50,218 egresos. No se detectaron registros con estadía negativa (los < 0), por lo que no fue necesario eliminar filas por inconsistencias.

La estadía promedio fue de 6.59 días, mientras que la mediana fue de 3 días, lo que sugiere una distribución asimétrica con presencia de casos de larga duración. El máximo observado fue de 614 días.

### B.0.5 Join diagnoses to CIE-10

```python
cie10_path = "/Users/ignaciorodriguezperez/Desktop/Preparacion_datos/CIE-10.xlsx"

cie10 = pd.read_excel(
    cie10_path,
    sheet_name="CIE 10"
)

cie10.columns = cie10.columns.str.strip()

print("Columnas en CIE-10:")
print(cie10.columns.tolist())

grd_com["DIAGNOSTICO1"] = grd_com["DIAGNOSTICO1"].astype(str).str.strip()
cie10["Código"] = cie10["Código"].astype(str).str.strip()

cie10_lookup = cie10[["Código", "Descripción", "Capítulo"]].drop_duplicates(subset="Código")

grd_com = grd_com.merge(
    cie10_lookup,
    left_on="DIAGNOSTICO1",
    right_on="Código",
    how="left"
)

# Renombrar columnas para trabajar más cómodo
grd_com = grd_com.rename(columns={
    "Código": "Codigo",
    "Descripción": "Descripcion",
    "Capítulo": "Capitulo"
})

n_no_match = grd_com["Capitulo"].isna().sum()
pct_no_match = n_no_match / len(grd_com) * 100

print(f"\nRows that failed to match ICD-10 lookup: {n_no_match:,}")
print(f"Percent unmatched: {pct_no_match:.2f}%")

print("\nSample unmatched DIAGNOSTICO1 values:")
print(
    grd_com.loc[grd_com["Capitulo"].isna(), "DIAGNOSTICO1"]
    .value_counts()
    .head(20)
)
```
